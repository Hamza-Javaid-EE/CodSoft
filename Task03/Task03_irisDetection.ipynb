##**Importing Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
```

##**Loading the Iris dataset**

```python
# Load the Iris dataset
iris = pd.read_csv('/content/IRIS.csv')
```

## **Displaying the first few rows of the dataset**

```python
print("First few rows of the dataset:")
print(iris.head())
```

## **Displaying statistical summary of the dataset**

```python
print("\nStatistical summary of the dataset:")
print(iris.describe())
```

## **Displaying unique target labels**

```python
print("\nUnique target labels:", iris["species"].unique())
```

## **Visualize the data using Seaborn pairplot**

```python
sns.set(style="ticks", color_codes=True)
g = sns.pairplot(iris, hue="species", markers=["o", "s", "D"])
g.fig.suptitle("Iris Flower Measurements", y=1.02)
plt.show()
```

## **Preparing the features (x) and target labels (y)**

```python
x = iris.drop("species", axis=1)
y = iris["species"]
```

## **Splitting the dataset into training and testing sets**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

## **Creating a K-Nearest Neighbors classifier with k=1**

```python
knn = KNeighborsClassifier(n_neighbors=3)  # You can adjust the number of neighbors
knn.fit(x_train, y_train)
```

## **Making a prediction using the trained classifier**

```python
x_new = pd.DataFrame([[6.3, 3.0, 5.5, 2.1]], columns=x.columns)
prediction = knn.predict(x_new)
print("\nPrediction:", prediction[0])
```

## **Evaluate the model**

```python
y_pred = knn.predict(x_test)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
